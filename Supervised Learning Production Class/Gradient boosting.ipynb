```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

np.random.seed(42)
x=np.random.rand(200,1)-0.5
y=3*x[:,0]**2+0.05*np.random.randn(200)

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)

gbrt = GradientBoostingRegressor(
    n_estimators=120,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

gbrt.fit(x_train, y_train)
```

```python
test_r2 = gbrt.score(x_test, y_test)
print("--- Gradient Boosting Production Metrics ---")
print(f"R² Performance Score on Test Dataset: {test_r2:.4f}")

X_grid = np.linspace(-0.5, 0.5, 500).reshape(-1, 1)
y_pred = gbrt.predict(X_grid)

plt.scatter(x, y, color="blue", alpha=0.5, label="Data Points")
plt.plot(X_grid, y_pred, color="red", linewidth=2, label="GBR Prediction Line")
plt.title("Gradient Boosting Regressor (GBRT) Predictions")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
