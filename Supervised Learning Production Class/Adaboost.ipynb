```python
from sklearn.datasets import make_moons
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text

x,y = make_moons(
    n_samples=100, 
    noise=0.25, 
    random_state=42
)
x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size=0.3, 
    random_state=42
)
```

```python
base_stump = DecisionTreeClassifier(max_depth=1, random_state=42)

ada_clf = AdaBoostClassifier(
    estimator=base_stump, 
    n_estimators=300, 
    learning_rate=0.2, 
    random_state=47
)

ada_clf.fit(x_train, y_train)

test_acc = ada_clf.score(x_test, y_test)
print("--- AdaBoost Ensemble Metrics ---")
print(f"Final Unseen Test Dataset Accuracy: {test_acc * 100:.2f}%")

print("\n--- Feature Importance Summary ---")
print(f"Feature 0 Importance: {ada_clf.feature_importances_[0]*100:.2f}%")
print(f"Feature 1 Importance: {ada_clf.feature_importances_[1]*100:.2f}%")
```

```python
stump_0 = ada_clf.estimators_[0]
print("--- 🗺️ Structural Map: AdaBoost Stump #0 ---")
print(export_text(stump_0, feature_names=["Feature_0", "Feature_1"]))

print("-" * 40)

stump_50 = ada_clf.estimators_[50]
print("--- 🗺️ Structural Map: AdaBoost Stump #50 ---")
print(export_text(stump_50, feature_names=["Feature_0", "Feature_1"]))
```
