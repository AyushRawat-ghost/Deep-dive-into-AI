```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

x,y=load_digits(
    # return_X_cube=False,
    return_X_y=True
)

x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
preprocessing_pipleine = Pipeline([ 
                                ('scaler',StandardScaler()),
                                ('kmeans_features',KMeans(random_state=42)),
                                ('log_reg',LogisticRegression(max_iter=5000))
                                ])

param_grid = {
    'kmeans_features__n_clusters':[40,50,60,70,80],
}
grid_search = GridSearchCV(
    preprocessing_pipleine,
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=1,
    scoring='accuracy'
)
grid_search.fit(x_train,y_train)
print("--- K-Means as a Preprocessing Step Results ---")
print(f"Optimal number of clusters chosen (K): {grid_search.best_params_['kmeans_features__n_clusters']}")
print(f"Pipeline Accuracy on Test Dataset    : {grid_search.score(x_test, y_test) * 100:.2f}%")
```
