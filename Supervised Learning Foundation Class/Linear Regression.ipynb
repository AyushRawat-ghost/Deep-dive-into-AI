```python
# Linear regression for marks prediction 
import numpy as np
import matplotlib .pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(42)
x=2*np.random.rand(100,1)
y=40+25*x+np.random.rand(100,1)*5
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x,y)

intercept=lin_reg.intercept_[0]
slope=lin_reg.coef_[0][0]

print("--- Trained Model Performance Parameters ---")
print(f"Intercept (Base Score when hours=0) : {intercept:.2f}")
print(f"Slope (Points gained per study hour): {slope:.2f}")
print(f"\nFinal Learned Equation: Score = {intercept:.2f} + ({slope:.2f} * Hours)")
```

```python
x_new=np.array([[1.5]])
predicted_score=lin_reg.predict(x_new)[0][0]
print(f"\nPredicted Score for 1.5 hours of study: {predicted_score:.2f}")

plt.scatter(x, y, color="blue", label="Actual Student Marks")
plt.plot(x, lin_reg.predict(x), color="red", linewidth=2, label="Linear Regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.legend()
plt.show()
```

```python
# Linear regression with gradient descent
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
m=1000
x1=2*np.random.rand(m,1)
x2=3*np.random.rand(m,1)
x3=np.random.rand(m,1)

x=np.c_[x1,x2,x3]
y=4+3*x1+1.5*x2+5*x3+np.random.rand(m,1)*0.2

scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

sgd_reg=SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.01,random_state=42)
sgd_reg.fit(x_scaled, y.ravel())
print("--- Learned Multivariable Parameters ---")
print(f"Intercept (Baseline Price) : {sgd_reg.intercept_[0]:.2f}")
print(f"Weight for Feature 1 (Size) : {sgd_reg.coef_[0]:.2f}")
print(f"Weight for Feature 2 (Rooms): {sgd_reg.coef_[1]:.2f}")
print(f"Weight for Feature 3 (Age)  : {sgd_reg.coef_[2]:.2f}")

X_new = np.array([[1.2, 2.0, 0.5]]) 
X_new_scaled = scaler.transform(X_new) 

predicted_price = sgd_reg.predict(X_new_scaled)[0]
print(f"\n🔮 Predicted Price for custom house: {predicted_price:.2f}")
```
