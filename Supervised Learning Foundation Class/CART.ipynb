```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=3, criterion="gini", random_state=42)
tree_clf.fit(X, y)

tree = tree_clf.tree_

print("--- Deep Structural Inspection of the CART Binary Graph ---")
print(f"Total number of nodes generated: {tree.node_count}")

for node_id in range(tree.node_count):
    is_leaf = tree.children_left[node_id] == -1
    
    print(f"\n🔹 Node ID {node_id}:")
    print(f"  ├── Samples passing through: {tree.n_node_samples[node_id]}")
    print(f"  ├── Computed Gini Impurity : {tree.impurity[node_id]:.4f}")
    
    if not is_leaf:
        feature_idx = tree.feature[node_id]
        feature_name = iris.feature_names[feature_idx]
        threshold = tree.threshold[node_id]
        
        print(f"  ├── [Internal Split Node]")
        print(f"  └── Action: Split if '{feature_name}' <= {threshold:.2f}")
        print(f"      ├── Left Child Target Node  ➔ ID {tree.children_left[node_id]}")
        print(f"      └── Right Child Target Node ➔ ID {tree.children_right[node_id]}")
    else:
        class_distributions = tree.value[node_id][0]
        predicted_class = np.argmax(class_distributions)
        predicted_name = iris.target_names[predicted_class]
        
        print(f"  ├── [Terminal Leaf Node]")
        print(f"  ├── Distribution: {class_distributions} (Class counts)")
        print(f"  └── Final Class Output Prediction: {predicted_name}")
```
