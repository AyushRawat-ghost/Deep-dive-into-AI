```python
import numpy as np
import matplotlib .pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

iris=load_iris()
x=iris["data"][:,3:]
y=(iris["target"]==2).astype(int)

log_reg=LogisticRegression(solver='lbfgs',random_state=42)
log_reg.fit(x,y)

x_new=np.linspace(0,3,1000).reshape(-1,1)
y_proba=log_reg.predict_proba(x_new)

decision_boundary = x_new[y_proba[:, 1] >= 0.5][0][0]
print(f" The learned decision boundary is at petal width: {decision_boundary:.2f} cm")

plt.plot(x_new, y_proba[:, 1], "g-", linewidth=2, label="Iris Virginica Probability")
plt.plot(x_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris Virginica")
plt.axvline(x=decision_boundary, color="red", linestyle=":", label="Decision Boundary")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Probability")
plt.title("Logistic Regression Sigmoid Curve on Iris Data")
plt.legend()
```
